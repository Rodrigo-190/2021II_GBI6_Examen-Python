## Ejercicio 4 [2 puntos]

Realice la visualización de los datos del ejercicio 3. 

* ```pubyear```: utilizando el conteo artículos por año, calcule la media y la desviación estandar (asumiendo que la distribución es normal). Realice un gráfico de la distribución normal de sus datos. Coloque el título, leyenda de los ejes "x" e "y", y la leyenda de la curva (utilice las expresiones matemáticas tipos latex de ser necesario). 

* ```pub_authors```: realice un boxplot del conteo de autores por artículo. Coloque las leyendas adecuadas para cada eje y el título.  

* ```countries```: realice una nube de puntos (adaptación de su tarea), resalte al menos 3 países con flechas. Coloque las leyendas adecuadas para cada eje y el título. 


````python
# Escriba aquí su código para el literal 'a' correspondiente al ejercicio 4

import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import csv as csv

print ('Distribución normal')
i = pub_year.iloc[:,1]
s = pd.Series(i)
h = list(set(s))
g=s
f=h
h=[0]*len(f)
k=0
for elem in f:
    d=0
    for comp in g:
        if elem == str(comp):
            d=d+1
    h[k]=d
    k=k+1
print ('Articulos por año :',h)
mean = np.average(h)
print ('El promedio es: ',mean)
stan = np.std(h)
print ('La desviación estándar es: ',stan)
x = np.linspace(mean - 3*stan, mean + 3*stan, 100)
sigma = stan
label = '$\mu = %.1f, \ \sigma^2=%.1f$' %(mean, stan)
y_norm = sts.norm.pdf(x, mean, sigma)
plt.plot(x, y_norm, label=label, c = 'lightseagreen')
plt.xlabel('Número de publicaciones')
plt.ylabel('Probabilidad de número de publicaciones')
plt.title('Distribución Normal')
plt.legend(title = "Parámetros")
plt.savefig("Distribución normal de la data.jpg")
plt.show()

````

````python
# Escriba aquí su código para el literal 'c' correspondiente al ejercicio 4
import re
import itertools
import csv
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

f = open('data/Ecuador genomics.txt',errors = 'ignore')
files = f.read()
File = re.sub(r"\n\s{6}", " " , files)

My_files = re.findall(r"\,\s[A-Z]+[A-z]{2,11}\.", File)
unique_codes = list(set(My_files))
unique_codes.sort()

a = "".join(unique_codes)
b = a.split(", ")
c = "".join(b)
unique_zipcodes = c.split(".")

#####cambiar Paises_ubicacion.csv #### por Coordenadas.csv
Coordenadas = pd.read_csv('data/data_Coordenadas.csv', engine='python', skipfooter=3)
Coordenadas.columns = ["ID_País", "Latitud", "Longitud", "Nombre"]

datos_nombre = Coordenadas[Coordenadas.Nombre.isin(unique_zipcodes)]
datos_paises = Coordenadas[Coordenadas.ID_País.isin(unique_zipcodes)]
all_data =pd.concat([datos_nombre,datos_paises]).drop_duplicates().sort_values('Nombre')

plt.scatter(all_data.Longitud, all_data.Latitud, s=100, c=all_data.Latitud)
plt.colorbar()
ard = dict(arrowstyle="->")
plt.annotate('AUSTRALIA', xy = (133.775136, -25.274398), 
               xytext = (133.775136, -20), arrowprops= ard)
plt.annotate('BRASIL', xy = (-51.92528, -14.235004), 
               xytext = (-51.92528, -9.5), arrowprops= ard)
plt.annotate('BELGICA', xy = (4.469936, 50.503887), 
               xytext = (9.469936, 38), arrowprops = ard)
plt.annotate('CHINA', xy = (104.195397, 35.86166), 
               xytext = (104.195397, 39), arrowprops= ard)
plt.annotate('NORUEGA', xy = (8.468946, 60.472024), 
               xytext = (8.468946, 63), arrowprops= ard)
plt.annotate('ARGENTINA', xy = (-63.616672,	-38.416097), 
               xytext = (-63.616672, -36), arrowprops= ard)
plt.annotate('BOLIVIA', xy = (-63.588653,	-16.290154), 
               xytext = (-63.588653, -6), arrowprops= ard)
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0] * 3, plSize[1] * 3) )
plt.title("Nube de puntos de publicaciones")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.savefig("Mapa de paises.jpg")
````
